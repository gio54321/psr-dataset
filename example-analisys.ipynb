# 2 year pulsar recording dataset
In this repository, apart from this example notebook, there is a big dataset containing (almost) daily observations of the pulsar PSR0329+54 at his culmination. The records are updated at 17/4/2018 but you can see live data coming out every day at [iw5bhy](iw5bhy.altervista.com) site.
The files in the dataset are
* data/index.txt: contains all the informations about the dataset in data/ (see below to get an idea)
* data/norm_profile.txt : the profile that is registered at ... for this pulsar
* data/[year]-[month]-[day]/profile.txt : profile file for that particular day (1d array)
* data/[year]-[month]-[day]/waterfall.txt : waterfall file for that particular day (2d array)



# example of elaboration
First we import all the libraries needed
* pandas: data analisys
* pyplot: plotting data
* numpy: numerical computation in python

then we show how index.txt (imported as a csv) looks like

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pd.set_option('display.max_rows', 20)

index = pd.read_csv('index.txt', sep=' ')

index
```

Now we want to create an array containing all the dates as np.datetime64 format. To do that we concatenate year, month and day to obtain a single string array that we convert to datetime64.

```python
days = np.core.defchararray.zfill(index['day'].as_matrix().astype('str'), 2)
months = np.core.defchararray.zfill(index['month'].as_matrix().astype('str'), 2)
years = index['year'].as_matrix().astype(str)
separator = np.chararray(len(days))
separator[:] = '-'

date_a = np.core.defchararray.add(years, separator.astype('str'))
date_b = np.core.defchararray.add(date_a, months)
date_c = np.core.defchararray.add(date_b, separator.astype('str'))
dates = np.core.defchararray.add(date_c, days)

dates = dates.astype('datetime64')
```

First of all let's see how peak varies in the long term period

```python
plt.rcParams['figure.figsize'] = (15, 7)

plt.plot(dates, index['peak'].as_matrix())
plt.title('Long term peak variation')
plt.xlabel('day')
plt.ylabel('peak(mdB)')
plt.grid(True)
plt.show()
```

now how signal to noise ratio varies in the long period

```python
plt.plot(dates, index['s/n'].as_matrix())
plt.title('Long term signal/noise variation')
plt.xlabel('day')
plt.ylabel('S/N')
plt.grid(True)
plt.show()
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    correction1 = np.empty(np.floor(box_pts/2).astype('int'))
    correction2 = np.empty(np.ceil(box_pts/2).astype('int')-1)
    correction1[:] = np.nan
    correction2[:] = np.nan
    y_smooth = np.append(correction1, np.convolve(y, box, mode='valid'))
    y_smooth = np.append(y_smooth, correction2)
    return y_smooth
```

```python
plt.subplot(2,1,1)
plt.plot(dates, index['peak'].as_matrix())
plt.plot(dates, smooth(index['peak'].as_matrix(), 50))
plt.title('Smoothed peak and S/N')
plt.ylabel('peak(mdB)')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(dates, index['s/n'].as_matrix())
plt.plot(dates, smooth(index['s/n'].as_matrix(), 50))
plt.xlabel('day')
plt.ylabel('S/N')
plt.grid(True)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(dates, index['peak'].as_matrix())
plt.plot(dates, smooth(index['peak'].as_matrix(), 50))
plt.title('Peak and S/N versus start time')
plt.ylabel('peak(mdB)')
plt.grid(True)
plt.xticks(dates[::10], index['start-time'][::10], size='small', rotation='vertical')

plt.subplot(2,1,2)
plt.plot(dates, index['s/n'].as_matrix())
plt.plot(dates, smooth(index['s/n'].as_matrix(), 50))
plt.xlabel('start-time')
plt.ylabel('S/N')
plt.grid(True)
plt.xticks(dates[::10], index['start-time'][::10], size='small', rotation='vertical')

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(dates, index['peak'].as_matrix())
plt.plot(dates, smooth(index['peak'].as_matrix(), 50))
plt.title('Peak and S/N versus end time')
plt.ylabel('peak(mdB)')
plt.grid(True)
plt.xticks(dates[::10], index['end-time'][::10], size='small', rotation='vertical')

plt.subplot(2,1,2)
plt.plot(dates, index['s/n'].as_matrix())
plt.plot(dates, smooth(index['s/n'].as_matrix(), 50))
plt.xlabel('end-time')
plt.ylabel('S/N')
plt.grid(True)
plt.xticks(dates[::10], index['end-time'][::10], size='small', rotation='vertical')

plt.show()
```
